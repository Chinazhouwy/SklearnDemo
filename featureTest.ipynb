```python
import categories
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats  # 导入 scipy.stats 模块

# 配置中文字体（以SimHei为例，根据系统调整）
plt.rcParams['font.sans-serif'] = ['Kai']  # 替换为系统中的中文字体名
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题
```

```python
df = pd.read_csv('stock_data/001222.SZ_2022-06-22_2025-06-21.csv')
print(df.head())
```

```python
# 假设 df 是包含源飞宠物（001222）日收益率的 DataFrame
sns.kdeplot(
    data=df,
    x="open",  # 日收益率列
    fill=True,
    color="purple",
    label="源飞宠物日收益率分布"
)

# 添加正态分布曲线对比
x = np.linspace(df["open"].min(), df["open"].max(), 100)
y = stats.norm.pdf(x, loc=df["open"].mean(), scale=df["open"].std())
plt.plot(x, y, linestyle="--", color="red", label="正态分布（同均值/标准差）")

plt.title("源飞宠物日收益率分布 vs 正态分布", fontsize=12)
plt.legend()
plt.show()
```
